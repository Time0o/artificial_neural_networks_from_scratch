# Hopfield Networks 4

Investigate noise resistance of Hopfield networks.

```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)

from ann.hopfield_network import HopfieldNetwork, profile_noise_resistance
```

# Load Images

```python
pict = np.genfromtxt('../data/pict.dat', delimiter=',').reshape(11, 1024)
```

# Train Network

```python
model = HopfieldNetwork()

model.train(pict[:3, :]);
```

# Profile Noise Resistance

```python
profile_noise_resistance(model, [pict[i, :] for i in range(3)])

plt.tight_layout()

plt.savefig('../figures/noise_resistance.svg')
```
