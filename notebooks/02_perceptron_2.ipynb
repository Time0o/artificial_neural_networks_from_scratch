# Perceptron 2

This notebook contains preliminary experiments testing the behaviour of the Perceptron rule for datasets that are not linearly separable.

```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(0)

from ann.dataset import create_dataset, plot_dataset
from ann.perceptron import perceptron_plot_generalization
from ann.plotting import savefig
from ann.subsampling import Subsampling, SUBSAMPLINGS
```

# Create Second Dataset

Create non-separable dataset.

```python
N = 100

LABEL_A = -1
LABEL_B = 1
```

```python
patterns2, targets2 = create_dataset(
    N, [((1, 0.3), 0.2), ((0, -0.1), 0.3)], [LABEL_A, LABEL_B], mirror=[True, False])

plot_dataset(patterns2, targets2)
```

# Subsampling Effects

Create different training datasets by imbalanced subsampling and visualize the effect this has on the resulting Perceptron rule decision boundary for several runs of the algorithm.

```python
perceptron_plot_generalization(patterns2, targets2, SUBSAMPLINGS, iterations=100, runs=50)

savefig('subsampling.svg')
```
